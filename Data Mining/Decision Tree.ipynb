```r
library(rpart)
library(rpart.plot)
d <- data.frame(y=sample(0:1,1000,repl=T), age=rnorm(1000,45,3.2),gender=sample(0:1,100, repl=T))
str(d)
```

```r
d$gender = as.factor(d$gender)
d$occu <- sample(1:4,1000,repl=T)
d$occu <- as.factor(d$occu)
levels(d$occu) <-c("D","E","B","L")
```

```r
rpart(y~.,data=d) -> dt
rpart(y~.,method="class",data=d) -> dt
rpart.plot(dt)
```

0-> label

.46 -> percentage of one's

98% -> part of total data
