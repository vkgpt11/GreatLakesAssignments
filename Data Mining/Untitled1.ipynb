# Hierarchical Clustering 

```r
dist(d1) -> dt
```

```r
hclust(dt,method  = "single") -> hsin
```

```r
plot(hsin)
```

```r
plot(hsin, hang=1)
```

```r
hclust(dt,method  = "complete") -> hcomp
plot(hcomp)
```

```r
hclust(dt,method  = "complete") -> havg
plot(havg)
```

```r
cutree(hcomp,4) -> hcom4
```

```r
hcom4
```

```r
d$hcluster <- hcom4
```

```r
d
```

```r
boxplot(height~hcluster, data=d)
```

```r
boxplot(age~hcluster, data=d)
```

```r
boxplot(inc~hcluster, data=d) 
```

```r
aov(inc ~ as.factor(hcluster),data=d) ->ft
summary(ft)
```

```r
aov(age ~ as.factor(hcluster),data=d) ->ft
summary(ft)
```

```r
aov(height ~ as.factor(hcluster),data=d) ->ft
summary(ft)
```

## Mixed Data set

```r
d2 <- data.frame(age= rnorm(10,34,2),gender=sample(0:1,10,repl=T),cal=sample(1:4,10,repl=T),satis=sample(1:5,10,repl=T))
```

```r
str(d2)
```

```r
d2$gender <- as.factor(d2$gender)
d2$cal <- as.factor(d2$cal)
d2$satis <- as.ordered(d2$satis) 
```

```r
str(d2)
```

```r
daisy(d2,metho,metric="gower") -> dt1
```

```r
dt1
```

```r
hclust(dt1,method="single") -> hclustmixed
plot(hclustmixed)
```
